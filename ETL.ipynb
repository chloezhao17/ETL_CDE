```python
import pandas as pd
from api_keys import g_key,ws_key
import requests
from pprint import pprint
```

```python
# Whole Foods Call Using Google API
url="https://maps.googleapis.com/maps/api/place/textsearch/json?"
params={
    "input":"Whole Foods",
    "key":g_key,
    "query":"Whole Foods in Georgia"
#     "fields":"formatted_address,name,geometry"
}
```

```python
response=requests.get(url,params=params).json()
# print(response.url)
# print(response.json())
```

```python
response
```

```python
data_ls=[]
for data in response["results"]:
        temp={
            "name":data["name"],
            "address":data["formatted_address"],
            "lat":data["geometry"]["location"]["lat"],
            "lng":data["geometry"]["location"]["lng"]            
        }
        data_ls.append(temp)
print(data_ls)
```

```python
wf_df=pd.DataFrame(data_ls)
wf_df
```

```python
wf_df.drop([11,13],inplace=True)
wf_df
```

```python
# Kroger Call Using Google API
url="https://maps.googleapis.com/maps/api/place/textsearch/json?"
params={
    "key":g_key,
    "query":"Kroger in Georgia"
}
```

```python
response=requests.get(url,params=params).json()
```

```python
data_ls=[]
for data in response["results"]:
        temp={
            "name":data["name"],
            "address":data["formatted_address"],
            "lat":data["geometry"]["location"]["lat"],
            "lng":data["geometry"]["location"]["lng"]            
        }
        data_ls.append(temp)
print(data_ls)
```

```python
k_df=pd.DataFrame(data_ls)
k_df
```

```python
df=wf_df.append(k_df)
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
# Walkscores initial pull
url="http://api.walkscore.com/score?"
ws=[]
for index, row in df.iterrows():
    lat=row["lat"]
    lng=row["lng"]
    address=row["address"]
    params={
        "format":"json",
        "address":address,
        "lat":lat,
        "lon":lng,
        "wsapikey":ws_key,
        "bike":1
    }
    data=requests.get(url,params=params).json()
    
#     for data in response:
#         print(data)
    try:
        temp={
            "walkscore":data["walkscore"],
            "ws description":data["description"],
#             "transit":data["transit"]["score"],
#             "t description":data["transit"]["description"],
            "bike":data["bike"]["score"],
            "b description":data["bike"]["description"],
            "address":address
        }
        ws.append(temp)
    except:
        pprint(data)
    
ws
```

```python
ws_df=pd.DataFrame(ws)
ws_df
```
